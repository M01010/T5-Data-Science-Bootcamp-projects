### video streaming

```python
import cv2
from ultralytics import YOLO
from ultralytics.utils.plotting import Annotator  # ultralytics.yolo.utils.plotting is deprecated

# Load YOLOv8 model
model = YOLO('new_best.pt')

video_path = "videoplayback.mp4"  # Specify the path to your video file
# Open video stream
cap = cv2.VideoCapture(video_path)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference
    results = model(frame)

    # Process detections
    for r in results:

        annotator = Annotator(frame)

        boxes = r.boxes
        for box in boxes:

            b = box.xyxy[0]  # get box coordinates in (left, top, right, bottom) format
            c = box.cls
            annotator.box_label(b, model.names[int(c)])

    frame = annotator.result()

    # Display frame
    cv2.imshow("Object Detection", frame)

    # Break loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture and close windows
cap.release()
cv2.destroyAllWindows()
```

quantization

```python
import torch
from torch.quantization import quantize_dynamic

# Load the YOLO model
model = YOLO('runs/detect/yolov8_Scoccer_Players_tuned/weights/best.pt')

# Apply post-training dynamic quantization
quantized_model = quantize_dynamic(model, dtype=torch.qint8)

# Save the quantized model
torch.save(quantized_model.state_dict(), 'quantized_yolo_model.pt')
```

pruning

```python
import torch.nn.utils.prune as prune

model = YOLO('runs/detect/yolov8_Scoccer_Players_tuned/weights/best.pt')

for name, layer in model.named_modules():
    if 'conv' in name:
        prune.l1_unstructured(layer, name='weight', amount=.2)


torch.save(model.state_dict(), 'pruned_yolo_model.pt')
```
