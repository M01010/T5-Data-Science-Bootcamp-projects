```python
encoding_map = {
    "sex" : {
        1 : "Male",
        0 : "Female"
        },
     'smoking': {
        0: 'Non-Smoking',
        1: 'Smoking',
        },
    "DEATH_EVENT":{
        0: "Alive",
        1: "Dead",
        },
      }
df_vis = df.replace(encoding_map)
```

```python
gender_colors = ['#d8849e', '#3c6ca8']
df_vis.groupby('sex').size().plot(kind='pie', autopct=lambda p: f'{p:.2f}%', colors=gender_colors)
plt.title('Distrubition of males and females')
plt.show()
```

```python
sns.histplot(x='age', data=df_vis, bins=11, color='#e3c414')
plt.title('Distrubition of age')
plt.show()
```

```python
smoke_colors = ['#68a358', '#b55c57']
df_vis.groupby('smoking').size().plot(kind='pie', autopct=lambda p: f'{p:.2f}%', colors=smoke_colors)
plt.title('Distrubition of smokers and non-smokers')
plt.show()

df_vis[df_vis.sex == 'Male'].groupby('smoking').size().plot(kind='pie', autopct=lambda p: f'{p:.2f}%', colors=smoke_colors)
plt.title('Distrubition of male smokers and non-smokers')
plt.show()

df_vis[df_vis.sex != 'Male'].groupby('smoking').size().plot(kind='pie', autopct=lambda p: f'{p:.2f}%', colors=smoke_colors)
plt.title('Distrubition of female smokers and non-smokers')
plt.show()
```

```python
ax = sns.barplot(x=['Diabetes', 'Anaemia', 'High Blood Pressure'], y=[len(df_vis[df_vis['diabetes'] == 1]),
len(df_vis[df_vis['anaemia'] == 1]),
len(df_vis[df_vis['high_blood_pressure'] == 1])], color='#636363')
plt.title('Number of affected people per CVD')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df_vis.groupby('DEATH_EVENT').size().plot(kind='pie', autopct=lambda p: f'{p:.2f}%', colors=['#636363', '#dddede'])
plt.title('Distrubition of dead and alive people')
plt.show()
```

```python
df_max = df.max()
df_norm = df.divide(df_max)

x = df_norm.drop(['DEATH_EVENT'], axis=1)
y = df_norm['DEATH_EVENT']
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=.2, random_state=0)
```

```python
from sklearn import linear_model, tree, svm, multiclass, naive_bayes

model = tree.ExtraTreeClassifier(random_state=0)
model.fit(x_train, y_train)
```

```python
from sklearn import metrics

y_pred = model.predict(x_test)

print(f'accuracy: {metrics.accuracy_score(y_test, y_pred)}')
print(f'precision: {metrics.precision_score(y_test, y_pred)}')
print(f'recall: {metrics.recall_score(y_test, y_pred)}')
print(f'mse: {metrics.mean_squared_error(y_test, y_pred)}')
```
