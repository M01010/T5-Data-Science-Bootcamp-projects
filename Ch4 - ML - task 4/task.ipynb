```python
import pandas as pd
import numpy as np
import os
from IPython.display import Audio
from transformers import VitsModel, AutoTokenizer
import torch
```

```python
model = VitsModel.from_pretrained("facebook/mms-tts-ara").cuda()
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-ara")


def get_speech(text):
    inputs = tokenizer(text, return_tensors="pt").to('cuda')

    with torch.no_grad():
        output = model(**inputs).waveform.cpu()
        return output
```

```python
data = get_speech("هذه تجربة. السلام عليكم")
Audio(data, rate=model.config.sampling_rate)
```

```python
df = pd.read_csv('data_ar_eg_test.tsv', sep='\t',
                 names=['id', 'path', 'transcription', 'raw_transcription', 'other', 'num_samples', 'gender'])
df = df[['path', 'transcription']]
```

```python
def get_audio(filepath):
    path = os.path.join('test', filepath)
    with open(path, 'rb') as f:
        return f.read()
```

```python
r_ind = np.random.choice(len(df))
sample = df.iloc[r_ind]

print(sample['transcription'])
Audio(get_audio(sample['path']))
```
