```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
df_reg = pd.read_csv('data/registration.csv')

df_train['train'] = 1
df_test['train'] = 0

df = pd.concat([df_train, df_test])
df
```

```python
df = df.merge(df_reg, how='left')
df
```

# filling null values

```python
# fix null values
cat_null_columns = ['Home Region', 'Home City', 'Employment Status']
num_null_columns = ['Age', 'University Degree Score System', 'University Degree Score']
columns_to_drop = ['College', 'Job Type', 'Still Working', 'Education Speaciality']

for c in cat_null_columns:
    df[c] = df[c].fillna(df[c].mode()[0])

for c in num_null_columns:
    df[c] = df[c].fillna(df[c].median())

for c in columns_to_drop:
    df.drop(c, inplace=True, axis=1)
```

```python
df.isna().sum()
```

# feature engineering

```python
df['Program Start Date'] = pd.to_datetime(df['Program Start Date'])
df['Program End Date'] = pd.to_datetime(df['Program End Date'])

df["Start Year"] = df["Program Start Date"].dt.year
df["Start Quarter"] = df["Program Start Date"].dt.quarter
df["Start Month"] = df["Program Start Date"].dt.month
df["Start Day"] = df["Program Start Date"].dt.day

df["End Year"] = df["Program End Date"].dt.year
df["End Quarter"] = df["Program End Date"].dt.quarter
df["End Month"] = df["Program End Date"].dt.month
df["End Day"] = df["Program End Date"].dt.day

df = df.drop(['Program End Date', 'Program Start Date'], axis=1)

df['Program Months'] = df['Program Days'].apply(lambda x: round(x / 30))

df['Weighted University Score'] = df['University Degree Score'] / df['University Degree Score System']

df = df.drop(['University Degree Score', 'University Degree Score System'], axis=1)


def filter_is_student(x):
    if 'طالب' in x:
        return 1
    return 0


def filter_is_working(x):
    if 'موظف' in x and x != 'غير موظف':
        return 1
    return 0


df['Is Student'] = df['Employment Status'].apply(filter_is_student)
df['Is Working'] = df['Employment Status'].apply(filter_is_working)

df = df.drop('Employment Status', axis=1)
```

```python
df
```

# encoding

```python
mapping_dict = {
    'Program Skill Level': {
        'مبتدئ': 0,
        'متوسط': 1,
        'متقدم': 2
    },
    'Level of Education': {
        'متوسط': 0,
        'ثانوي': 1,
        'الدبلوم': 2,
        'البكالوريوس': 3,
        'الماجستير': 4,
        'الدكتوراه': 5
    },
}
df = df.replace(mapping_dict)
```

```python
from sklearn.preprocessing import OrdinalEncoder

cols = ['Home Region', 'Home City', 'Program Main Category Code', 'Program Sub Category Code', 'Gender',
        'Program Presentation Method',
        'Completed Degree', 'Technology Type', 'Program ID']
encoder = OrdinalEncoder()

df[cols] = encoder.fit_transform(df[cols])
```

```python
from sklearn.impute import KNNImputer

numeric_cols = df.drop(['Student ID', 'Y'], axis=1).columns
df[numeric_cols] = KNNImputer().fit_transform(df[numeric_cols])
```

# Model building

```python
x = df.drop(['Student ID', 'Y'], axis=1)
y = df[['Y']]
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

x, y = SMOTE(random_state=42).fit_resample(x, y)

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
sns.displot(y)
plt.savefig('fig2.png')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

logistic = LogisticRegression(solver='liblinear', random_state=42)
logistic.fit(x_train, y_train.values.ravel())

tree = DecisionTreeClassifier(random_state=42)
tree.fit(x_train, y_train.values.ravel())
```

```python
from sklearn.metrics import f1_score

scores = []

y_pred = logistic.predict(x_test)
print(logistic.__class__.__name__)
print(f'f1: {f1_score(y_test, y_pred)}\n')
scores.append((logistic.__class__.__name__, f1_score(y_test, y_pred)))

y_pred = tree.predict(x_test)
print(tree.__class__.__name__)
print(f'f1: {f1_score(y_test, y_pred)}')
scores.append((tree.__class__.__name__, f1_score(y_test, y_pred)))
```

# Model improvement

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
import xgboost as xgb

models = [RandomForestClassifier(random_state=42),
          GradientBoostingClassifier(random_state=42),
          AdaBoostClassifier(random_state=42),
          xgb.XGBClassifier(random_state=42)
          ]

for model in models:
    model.fit(x_train, y_train.values.ravel())
```

```python
for model in models:
    y_pred = model.predict(x_test)
    print(model.__class__.__name__)
    print(f'f1: {f1_score(y_test, y_pred)}\n')
    scores.append((model.__class__.__name__, f1_score(y_test, y_pred)))
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'max_depth': [10, 20, 30, 40, None],
    'n_estimators': [250, 300, 350],
    'max_features': [2, 3, 5, 7, 'sqrt', 'log2'],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [3, 4, 5, 6]
}
random_model = RandomizedSearchCV(RandomForestClassifier(random_state=42), param_grid,
                                  cv=3, n_jobs=-1, random_state=42)
random_model.fit(x_train, y_train.values.ravel())
```

```python
random_model.best_params_
```

```python
best = RandomForestClassifier(random_state=42, **random_model.best_params_)
best.fit(x_train, y_train.values.ravel())

y_pred = best.predict(x_test)
print(f'f1: {f1_score(y_test, y_pred)}\n')
```

# ensemble of the best model

```python
from sklearn.ensemble import BaggingClassifier

ensemble_model = BaggingClassifier(RandomForestClassifier(random_state=42, **random_model.best_params_), random_state=42)
ensemble_model.fit(x_train, y_train.values.ravel())

y_pred = ensemble_model.predict(x_test)
print(f'f1: {f1_score(y_test, y_pred)}\n')
```
