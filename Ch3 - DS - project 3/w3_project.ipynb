```python
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Airbnb Data/Listings.csv', encoding='latin-1', low_memory=False)
print(df.shape)
```

```python
indexes = df[(df.price == 0) | (df.minimum_nights > 2000)].index
df.drop(indexes, inplace=True)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
#fix data types
df['host_since'] = pd.to_datetime(df['host_since'])
df['instant_bookable'] = df['instant_bookable'].apply(lambda x: 0 if x == 'f' else 1)
df['host_identity_verified'] = df['host_identity_verified'].apply(lambda x: 0 if x == 'f' else 1)
df['host_has_profile_pic'] = df['host_has_profile_pic'].apply(lambda x: 0 if x == 'f' else 1)
df['host_is_superhost'] = df['host_is_superhost'].apply(lambda x: 0 if x == 'f' else 1)
```

```python
# fix null values
cat_null_columns = ['name', 'host_location', 'host_response_time']
num_null_columns = ['host_since', 'bedrooms', 'host_response_rate', 'host_acceptance_rate', 'host_total_listings_count',
                    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
                    'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
                    'review_scores_value']
columns_to_drop = ['district']

df_filled = df.copy()

for c in cat_null_columns:
    df_filled[c] = df_filled[c].fillna(df_filled[c].mode()[0])

for c in num_null_columns:
    df_filled[c] = df_filled[c].fillna(df_filled[c].median())

for c in columns_to_drop:
    df_filled.drop(c, inplace=True, axis=1)
```

```python
df_filled['bedrooms'] = df_filled['bedrooms'].astype(int)
df_filled['host_response_time'] = df_filled['host_response_time'].map(
    {
        'within an hour': 0,
        'within a few hours': 1,
        'within a day': 2,
        'a few days or more': 3
    }
)
```

```python
df_reviews = pd.read_csv('Airbnb Data/Reviews.csv')
df_reviews = df_reviews.groupby('listing_id', as_index=False)['review_id'].count()
df_reviews.rename({'review_id': 'total_reviews'}, axis=1, inplace=True)
```

```python
df_reviews
```

```python
# feature engineering
df_filled = df_filled.merge(df_reviews, on='listing_id', how='left')
df_filled['total_reviews'] = df_filled['total_reviews'].fillna(0)
df_filled['host_since_year'] = df_filled['host_since'].dt.year
df_filled['host_since_month'] = df_filled['host_since'].dt.month
df_filled['host_since_year_month'] = df_filled['host_since'].apply(lambda d: datetime.datetime(d.year, d.month, 1))
```

```python
# Calculate the first quartile (Q1)
Q1 = df_filled.price.quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = df_filled.price.quantile(0.75)

# Calculate the IQR (Inter-quartile Range)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df_filled[(df_filled.price >= lower_bound) & (df_filled.price <= upper_bound)]
```
